```python
# Find the number of events within a distance of 1 from a given event
def euclidean_distance(x1, y1, z1, x2, y2, z2):
    return math.sqrt((x2 - x1)**2 + (y2 - y1)**2 + (z2 - z1)**2)

# Given event_id
given_event_id = '4090686'

# Find the position (x, y, z) of the given event
given_event = collection.find_one({"event_id": given_event_id})
given_x = float(given_event['x'])
given_y = float(given_event['y'])
given_z = float(given_event['z'])

# Find events within a distance of 1 from the given event
events_within_distance = []
for event in collection.find():
    event_x = float(event['x'])
    event_y = float(event['y'])
    event_z = float(event['z'])
    distance = euclidean_distance(given_x, given_y, given_z, event_x, event_y, event_z)
    if distance <= 1:
        events_within_distance.append(event)

print("Number of records:", len(events_within_distance))
```

```python
# Distribution of average charge
for document in collection.find():
    avg_charge_float = float(document["avg_charge"])
    collection.update_one(
        {"_id": document["_id"]},
        {"$set": {"avg_charge": avg_charge_float}}
    )
    
charge_ranges = [
    {"$gte": 0, "$lt": 3},
    {"$gte": 3, "$lt": 6},
    {"$gte": 6, "$lt": 9},
    {"$gte": 9, "$lt": 12},
]

for charge_range in charge_ranges:
    min_value = charge_range["$gte"]
    max_value = charge_range["$lt"]
    count = collection.count_documents({
        "$and": [
            {"avg_charge": {"$gte": float(min_value)}},
            {"avg_charge": {"$lt": float(max_value)}}
        ]
    })
    print(f"Range: {min_value} - {max_value}, Count: {count}")
```

```python
# Distribution of number of unique pulses

for document in collection.find():
    n_unique_pulses_float = float(document["n_unique_pulses"])
    collection.update_one(
        {"_id": document["_id"]},
        {"$set": {"n_unique_pulses": n_unique_pulses_float}}
    )
n_unique_pulses_ranges = [
    {"$gte": 0, "$lt": 20},
    {"$gte": 20, "$lt": 40},
    {"$gte": 40, "$lt": 60},
    {"$gte": 60, "$lt": 80}
]

# Iterate over each n_unique_pulses range and count events falling into that range
for range_query in n_unique_pulses_ranges:
    min_value = range_query["$gte"]
    max_value = range_query["$lt"]
    count = collection.count_documents({
        "$and": [
            {"n_unique_pulses": {"$gte": min_value}},
            {"n_unique_pulses": {"$lt": max_value}}
        ]
    })
    print(f"Range: {min_value} - {max_value}, Count: {count}")
```
